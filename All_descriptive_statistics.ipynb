```python
import pandas as pd
```

```python
excel_file1 = 'corrected_series_India.xlsx'    ### country
dfi = pd.read_excel(excel_file1)
dfi
```

```python
del dfi["Unnamed: 0"]
del dfi["Date"]
sti = dfi.describe().T
sti
```

```python
excel_file1 = 'corrected_series_China.xlsx'    ### country
dfc = pd.read_excel(excel_file1)
dfc
```

```python
del dfc["Unnamed: 0"]
del dfc["Date"]
stc = dfc.describe().T
stc
```

```python
excel_file1 = 'corrected_series_exog.xlsx'    ### country
dfe = pd.read_excel(excel_file1)
dfe
```

```python
del dfe["Unnamed: 0"]
del dfe["Date"]
ste = dfe.describe().T
ste
```

```python

with pd.ExcelWriter('DescriptiveStatisticsALLvkr.xlsx') as writer:
    sti.to_excel(writer, sheet_name='Descriptive Statistics India')
    stc.to_excel(writer, sheet_name='Descriptive Statistics China') 
    ste.to_excel(writer, sheet_name='Descriptive Statistics Exog') 
```
